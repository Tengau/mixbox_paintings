## Introduction
The Kubelka-Munk theory, developed by Paul Kubelka and Franz Munk, is one of the most successful models of mixing paint. Unlike RGB, RYB, Subtractive, and Subtractive-Additive models, the K-M model follows the actual mixing process of paints that occur in real life. Within the paint, incident light and pigment particles interact by having the particles absorb and scatter the light depending on the wavelength. Ultimately, the difference in these absorption and scattering coefficients cause different paints to have different colors. 

The K-M model implemented in this Jupyter Python 3 notebook is performed using the open source mixbox library. Provided with the RGB values of colors and the concentrations of the pigments, the RGB values for the mixture can be calculated.

## Generate a Palette - Painting 1

```python
import mixbox
import matplotlib.pyplot as plt
import numpy as np
%pip install seaborn

rgb_white = (255, 255, 255) # white
rgb_magenta = (128, 2, 46)  # Quinacridone Magenta
rgb_cadmium_yellow = (254, 236, 0) # Cadmium Yellow
rgb_cobalt_blue = (0, 33, 133) # Cobalt Blue

def generateRGB_image(RGB):
    height, width, channel = 64, 64, 3
    arr = np.full((height, width, channel), [RGB[0], RGB[1], RGB[2]], dtype=('uint8'))
    plt.imshow(arr)
    plt.axis('off')

# mixing multipule pigments
def mix(rgb1, rgb2, rgb3, rgb4, c):
    z1 = mixbox.rgb_to_latent(rgb1)
    z2 = mixbox.rgb_to_latent(rgb2)
    z3 = mixbox.rgb_to_latent(rgb3)
    z4 = mixbox.rgb_to_latent(rgb4)

    z_mix = [0] * mixbox.LATENT_SIZE

    for i in range(len(z_mix)):     # mix together:
        z_mix[i] = (c[0]*z1[i] +     #   % of rgb1
                    c[1]*z2[i] +     #   % of rgb2
                    c[2]*z3[i] +     #   % of rgb3
                    c[3]*z4[i])      #   % of rgb4

    rgb_mix_mult = mixbox.latent_to_rgb(z_mix)
    return rgb_mix_mult
```

```python
mixed_rgb_bkg_1 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.2, 0.5, 0.3, 0.0])
print(mixed_rgb_bkg_1)
generateRGB_image(mixed_rgb_bkg_1)
```

```python
mixed_rgb_bkg_2 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.4, 0.3, 0.3, 0.0])
print(mixed_rgb_bkg_2)
generateRGB_image(mixed_rgb_bkg_2)
```

```python
mixed_rgb1 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.1, 0.0, 0.0, 0.9])
print(mixed_rgb1)

generateRGB_image(mixed_rgb1)
```

```python
mixed_rgb2 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.75, 0.01, 0.04, 0.2])
print(mixed_rgb2)
generateRGB_image(mixed_rgb2)
```

```python
mixed_rgb3 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.0, 0.2, 0.2, 0.6])
print(mixed_rgb3)
generateRGB_image(mixed_rgb3)
```

```python
mixed_rgb4 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.9, 0.05, 0.05, 0.0])
print(mixed_rgb4)
generateRGB_image(mixed_rgb4)
```

```python
mixed_rgb5 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.8, 0.1, 0.0, 0.1])
print(mixed_rgb5)
generateRGB_image(mixed_rgb5)
```

## Conclusions

Throughout this process, I have learned a lot about color theory and possible applications of the mixbox library. Usually, when I paint with acrylic, I mix my color palette by eye. This method of mixing colors until the desired colors is mixed is something that I have learned through my experience painting. This method requires me to understand how paint changes color as it dries, how different tones and shade of colors can be created, and how the color wheel works to change the saturation, hue, and value of a color. For a beginner in working with color, this method may be difficult to adjust to. 

#### Usage 1: For Beginners
During this project, however, I used concentrations of paint to determine exactly how a color will appear after it has been mixed. Breaking down the usual experience-driven method into numbers and algorithms may be helpful for some people in understanding how mixing colors with paint works. This is a great tool for beginners to understand how mixing colors works in real life. 

#### Usage 2: For Time Constraints
When setting values for my paint mixtures, I was also able to experiment with many color combinations, that would have been difficult to implement on an actual canvas, due to time constraints. The ease of changing a value in a Jupyter Notebook and running a program is incomparable to mixing colors in real life. Additionally, using values in a program has the benefit of saving/reproducing every color combination you have created before. This is something especially important in acrylic painting, because of how quickly acrylic paint dries. Once your paint has dried, the entire process of color matching via eye has to be repeated.

Ultimately, I had a great time painting the subject and applying the palette constraints. As a result of using a set palette (7 or 8 colors mixed from 3 primary colors and white), the overall color scheme was cohesive within the painting.
