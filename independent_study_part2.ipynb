## Generate a Palette

```python
import mixbox
import matplotlib.pyplot as plt
import numpy as np
#%pip install seaborn

rgb_white = (255, 255, 255) # white
rgb_magenta = (128, 2, 46)  # Quinacridone Magenta
rgb_cadmium_yellow = (254, 236, 0) # Cadmium Yellow
rgb_cobalt_blue = (0, 33, 133) # Ultramarine Blue

def generateRGB_image(RGB):
    height, width, channel = 64, 64, 3
    arr = np.full((height, width, channel), [RGB[0], RGB[1], RGB[2]], dtype=('uint8'))
    plt.imshow(arr)
    plt.axis('off')

# mixing multipule pigments
def mix(rgb1, rgb2, rgb3, rgb4, c):
    z1 = mixbox.rgb_to_latent(rgb1)
    z2 = mixbox.rgb_to_latent(rgb2)
    z3 = mixbox.rgb_to_latent(rgb3)
    z4 = mixbox.rgb_to_latent(rgb4)

    z_mix = [0] * mixbox.LATENT_SIZE

    for i in range(len(z_mix)):     # mix together:
        z_mix[i] = (c[0]*z1[i] +     #   % of rgb1
                    c[1]*z2[i] +     #   % of rgb2
                    c[2]*z3[i] +     #   % of rgb3
                    c[3]*z4[i])      #   % of rgb4

    rgb_mix_mult = mixbox.latent_to_rgb(z_mix)
    return rgb_mix_mult
```

```python
mixed_rgb_bkg_1 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.3, 0.0, 0.15, 0.55])
print(mixed_rgb_bkg_1)
generateRGB_image(mixed_rgb_bkg_1)
```

```python
mixed_rgb_bkg_2 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.7, 0.00, 0.1, 0.2])
print(mixed_rgb_bkg_2)
generateRGB_image(mixed_rgb_bkg_2)
```

```python
mixed_rgb1 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.9, 0.1, 0.0, 0.0])
print(mixed_rgb1)

generateRGB_image(mixed_rgb1)
```

```python
mixed_rgb2 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.2, 0.01, 0.04, 0.75])
print(mixed_rgb2)
generateRGB_image(mixed_rgb2)
```

```python
mixed_rgb4 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.45, 0.5, 0.0, 0.05])
print(mixed_rgb4)
generateRGB_image(mixed_rgb4)
```

```python
mixed_rgb5 = mix(rgb_white, rgb_magenta, rgb_cadmium_yellow, rgb_cobalt_blue, [0.9, 0.00, 0.05, 0.05])
print(mixed_rgb5)
generateRGB_image(mixed_rgb5)
```

```python
import pandas as pd
import seaborn as sns

def create_color(rgb):
    r = rgb[0]
    g = rgb[1]
    b = rgb[2]
    return (r/256, g/256, b/256)

sns.set(rc={"figure.figsize":(6, 6)})
color_1 = create_color(mixed_rgb_bkg_1)
color_2 = create_color(mixed_rgb_bkg_2)
color_3 = create_color(mixed_rgb1)
color_4 = create_color(mixed_rgb2)
color_5 = create_color(mixed_rgb3)
color_6 = create_color(mixed_rgb4)
color_7 = create_color(mixed_rgb5)
color_8 = create_color(mixed_rgb6)

current_palette = {color_1, color_2, color_3, color_4, color_5, color_6, color_7, color_8}
sns.palplot(current_palette)
```
